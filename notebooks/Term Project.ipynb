```python
#standardize the features
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

features=Emission.drop(columns=['Make','Model','Vehicle Class','Transmission','Fuel Type','CO2 Emissions(g/km)'])
y=Emission['CO2 Emissions(g/km)']
scaler=StandardScaler()
scaled_features=scaler.fit_transform(features)
```

```python
#apply priincipal component analysis
pca=PCA(n_components=2)
transformed_data=pca.fit_transform(scaled_features)
print(pca.explained_variance_ratio_)
```

```python
#build a linear model and evaulate model using r_squared,mean_squared_error
#and RMSE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np

X_train,X_test,y_train,y_test=train_test_split(transformed_data,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(X_train,y_train)
linear_pred=model.predict(X_test)
r_squared=r2_score(y_test,linear_pred)
mse=mean_squared_error(y_test,linear_pred)
rmse=np.sqrt(mse)
print("R-squared score:",r_squared)
print("Mean-Squared_Error:",mse)
print("RMSE:",rmse)
```

```python
# Access regression coefficients. Identify which principal component had the largest influence 
# on CO2 Emissions
coefficients = pd.DataFrame(model.coef_, index=[f'PC{i+1}' for i in range(transformed_data.shape[1])], columns=['Coefficient'])

print(coefficients)
```

```python
# Create a DataFrame for easier interpretation of features that had the strongest impact on
# CO2 Emissions

pca_components = pd.DataFrame(
    pca.components_, 
    columns=features.columns,  
    index=[f'PC{i+1}' for i in range(pca.n_components_)]
)

print(pca_components)
```

```python
#Fuel Consumption City, fuel consumption hwy, and Fuel Consumption Comb contribute the most to CO2 Emissionsw
```

```python
#use lasso and ridge regression to identify important features that contribute to CO2 Emissions
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge,Lasso,RidgeCV,LassoCV
X_train2,X_test2,y_train2,y_test2=train_test_split(scaled_features,y,test_size=0.2,random_state=42)
alphas=[0.001,0.01,0.1,1,10,100]
```

```python
#obtain optimal alpha for ridge regression
ridge_cv=RidgeCV(alphas=alphas,cv=10)
ridge_cv.fit(X_train2,y_train2)
optimal_alpha=ridge_cv.alpha_
print(optimal_alpha)
```

```python
#learn a ridge model with optimal alpha and evaulate model using r_squared,mean_squared_error
#and RMSE
ridge_model=Ridge(alpha=optimal_alpha)
ridge_model.fit(X_train2,y_train2)
ridge_pred=ridge_model.predict(X_test2)
ridge_r_squared=r2_score(y_test2,ridge_pred)
ridge_mse=mean_squared_error(y_test2,ridge_pred)
ridge_rmse=np.sqrt(ridge_mse)

print("R-squared score:",ridge_r_squared)
print("Mean-Squared_Error:",ridge_mse)
print("RMSE:",ridge_rmse)
```

```python
# Display the most impactful features
ridge_coefficients = ridge_model.coef_
feature_importance = sorted(zip(ridge_coefficients, features), key=lambda x: abs(x[0]), reverse=True)
print("Feature Impact on CO2 Emissions:")
for coeff, feature in feature_importance:
    print(feature,":",coeff)
   
```

```python
#ridge regression indicates Fuel Consumption Comb(mpg) has the biggest impact on CO2 Emissions
```

```python
# finding the optimal alpha for Lasso regression
lasso_cv=LassoCV(alphas=None,cv=10)
lasso_cv.fit(X_train2,y_train2)
optimal_alpha2=lasso_cv.alpha_
print("optimal_alpha",optimal_alpha2)
```

```python
#learn a lasso model with optimal alpha and evaulate model using r_squared,mean_squared_error
#and RMSE
lasso_model=Lasso(alpha=optimal_alpha2)
lasso_model.fit(X_train2,y_train2)
lasso_pred=lasso_model.predict(X_test2)
lasso_r_squared=r2_score(y_test2,lasso_pred)
lasso_mse=mean_squared_error(y_test2,lasso_pred)
lasso_rmse=np.sqrt(lasso_mse)

print("R-squared score:",lasso_r_squared)
print("Mean-Squared_Error:",lasso_mse)
print("RMSE:",lasso_rmse)
```

```python
# Display the most impactful features
lasso_coefficients = lasso_model.coef_
feature_importance = sorted(zip(lasso_coefficients, features), key=lambda x: abs(x[0]), reverse=True)
print("Feature Impact on CO2 Emissions:")
for coeff, feature in feature_importance:
    print(feature,":",coeff)
```
